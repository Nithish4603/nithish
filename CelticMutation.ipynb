```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
#loading the data
df = pd.read_csv('/content/train.tsv', sep='\t',header= None,encoding='utf8')
```

```python
df.head()
```

```python
df.columns = ['text','tag']
df
```

```python
#for loop to add the elements
token = []
for i in range(1,df.shape[0]):
  x =  str(df['text'][i-1]) + ' ' + str(df['text'][i] )
  token.append(x)
```

```python
df['text'] = pd.DataFrame (token, columns = ['text'])
df
```

```python
#Creating a list for df['tag'] and removing the first elemnent
listt = df['tag'].tolist()
listt = listt[1:]
```

```python
# Dropping the last index from df
df.drop(df.tail(1).index,inplace=True)
```

```python
#Convertings list to df['tags']
df['tags'] = listt
```

```python
df
```

```python
# Removing <S>
df['text'] = df['text'].str.replace('<S>','')
```

```python
# Removing <NUM>
df['text'] = df['text'].str.replace('<NUM>','')
```

```python
# Removing <URL>
df['text'] = df['text'].str.replace('<URL>','')
```

```python
# Removing <EMAIL>
df['text'] = df['text'].str.replace('<EMAIL>','')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
df['tags'] = encoder.fit_transform(df['tags'].fillna(' '))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=1000)
cv = vectorizer.fit_transform(df['text'])
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cv,df['tags'], test_size=0.2, random_state=0)
```

```python
#Training the data
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
