```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk
```

```python
df = pd.read_csv('/content/train.tsv', sep='\t',header= None,encoding='utf8')
```

```python
df.head()
```

```python
df.columns = ['text','tag']
df
```

```python
prec_token = []
for i in range(1,df.shape[0]):
  k =  str(df['text'][i-1]) + ' ' + str(df['text'][i] )
  prec_token.append(k)
```

```python
df['text'] = pd.DataFrame (prec_token, columns = ['text'])
df
```

```python
listt = df['tag'].tolist()
listt = listt[1:]
```

```python
df.drop(df.tail(1).index,inplace=True)
```

```python
df['tags'] = listt
```

```python
df
```

```python
df['text'] = df['text'].str.replace('<S>','')
```

```python
df['text'] = df['text'].str.replace('<NUM>','')
```

```python
df['text'] = df['text'].str.replace('<URL>','')
```

```python
df['text'] = df['text'].str.replace('<EMAIL>','')
```

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
df['tags'] = en.fit_transform(df['tags'].fillna(' '))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(max_features=1000)
cv = vec.fit_transform(df['text'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cv,df['tags'], test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
